```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook

import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("data/clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data.head()
clinical_trial_data.head()

# Combine the data into a single dataset
combined_trial_data = pd.merge(clinical_trial_data,mouse_drug_data, on="Mouse ID" , how='left')


# Display the data table for preview
combined_trial_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
group_mean=combined_trial_data.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)']
group_mean
# Convert to DataFrame

group_mean_data=pd.DataFrame(group_mean).reset_index()


# Preview DataFrame
group_mean_data
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

group_sem=combined_trial_data.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)']

# Convert to DataFrame

group_sem_data=pd.DataFrame(group_sem).reset_index()


# Preview DataFrame

group_sem_data

```

```python
# Minor Data Munging to Re-Format the Data Frames
reset_index_group= group_mean_data.reset_index()
reformat_mean=reset_index_group.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
reformat_mean
reformat_sem= group_sem_data.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']

# Preview that Reformatting worked
reformat_mean.head()


```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0,50,10)



plt.errorbar(reformat_mean.index, reformat_mean["Capomulin"], yerr=reformat_sem["Capomulin"], color="r", marker="o", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(reformat_mean.index, reformat_mean["Infubinol"], yerr=reformat_sem["Infubinol"], color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(reformat_mean.index, reformat_mean["Ketapril"], yerr=reformat_sem["Ketapril"], color="g", marker="s", markersize=5, linestyle="dashed", linewidth=0.75)
plt.errorbar(reformat_mean.index, reformat_mean["Placebo"], yerr=reformat_sem["Placebo"], color="k", marker="d", markersize=5, linestyle="dashed", linewidth=0.75)


plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)") 
plt.ylabel("Tumor Volume (mm3)") 

plt.style.use('seaborn-whitegrid')
plt.grid(linestyle="dashed")

plt.grid(linestyle="dashed")
plt.legend(loc='best', fontsize=12)

# Save the figure
plt.savefig("TumerResponseTreatment.png")

# Show the Figure
plt.show()
```
